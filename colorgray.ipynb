```python
import cv2
img = cv2.imread('salman.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imGr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("RGB Image")
plt.subplot(1, 2, 2)
plt.imshow(imGr, cmap='gray', vmin = 0, vmax = 255,interpolation='none')
plt.title("Gray Scale Image")
plt.show()
```
